```python
import pandas as pd
drug=pd.read_csv('doses.csv')
drug.head()
```

```python
len(drug[drug['time_of_effect']<11.5])/len(drug)*100
```

```python
from scipy.stats import norm

mu=drug['time_of_effect'].mean()
sigma=drug['time_of_effect'].std()
print(round(mu,2))
print(round(sigma,2))
```

```python
plt.figure(figsize=(10,4))
sns.histplot(drug['time_of_effect'])
```

```python
drug['time_of_effect'].mean(),drug['time_of_effect'].median()
```

```python
import scipy.stats as stats
stats.norm.cdf(11.5,drug['time_of_effect'].mean(),drug['time_of_effect'].std())
```

```python
#A) probability that the time of effect of a dose is less than 11.5 hours
prob=norm.cdf(11.5,mu,sigma)
print('probability that the time of effect of a dose is less than 11.5 is',round(prob,4))

#B) probability tha the time of effect of a dose is more than 10 hours
prob1=1-norm.cdf(10,mu,sigma)
print('probability that the time of effect of a dose is more than 10 is',round(prob1,4))

#C) Calculate the 90th percentile of time of effect for the doses
print('90th percentile of time of effect is',round(norm.ppf(0.9,mu,sigma),2),'hrs')
```
